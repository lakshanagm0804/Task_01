```python
# loading data
import pandas as pd
df = pd.read_csv(r"E:\Task1\Dataset\netflixData_raw_data.csv")
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
# removing duplicates
dup = df.duplicated().sum()
print(dup)
```

```python
df = df.drop_duplicates()
```

```python
df.shape
```

```python
# Null handling
df.isnull().sum()
```

```python
df["Director"] = df["Director"].fillna("Unknown")
```

```python
df["Director"].isna().sum()
```

```python
df["Cast"] = df["Cast"].fillna("Not Mentioned")
```

```python
df["Cast"].isnull().sum()
```

```python
df["Production Country"] = df["Production Country"].fillna("Unknown")
```

```python
df["Production Country"].isnull().sum()
```

```python
df = df.dropna(subset="Release Date")
```

```python
df["Release Date"].isna().sum()
```

```python
df["Rating"] = df["Rating"].fillna("No Rating")
```

```python
df["Release Date"].isna().sum()
```

```python
df = df.dropna(subset="Duration")
```

```python
df["Duration"].isna().sum()
```

```python
df["Imdb Score"] = df["Imdb Score"].fillna("Not Available")
```

```python
df["Imdb Score"].isna().sum()
```

```python
df[df["Date Added"].isna()]
```

```python

df["Date Added"].dtype
```

```python
df["Date Added"] = df["Date Added"].bfill()
```

```python
df["Date Added"].isna().sum()
```

```python
df.isnull().sum()
```

```python
# Changing Datatypes
```

```python
df.dtypes
```

```python
df["Title"] = df["Title"].astype("string")
```

```python
columns = ["Description","Director" , "Genres", "Cast", "Production Country", "Content Type"]
```

```python
df[columns] = df[columns].astype("string")
```

```python
df["Date Added"] = pd.to_datetime(df["Date Added"])
```

```python
df.dtypes
```

```python
df.head()
```

```python
# Standardizing columns
df.colums = df.columns.str.replace("","_")
```

```python
df.head()
```

```python
df.to_csv(r"E:\Task1\Dataset\Netflix_cleaned_data.csv", index=False)
```
